```python
data=data.iloc[:,1:]
data.head()
```

```python
from sklearn.preprocessing import scale,normalize
```

```python
data1=scale(data)
```

```python
data1
```

```python
data2=normalize(data1)
```

```python
data2
```

```python
from sklearn.decomposition import PCA
```

```python
model=PCA()
```

```python
model.fit(data2)
```

```python
model.components_
```

```python
#convert 1D to 2D
componet=pd.DataFrame(model.components_)
```

```python
componet.head()
```

```python
value=model.explained_variance_ratio_
```
